### Listing all the columns that are currently non-numeric:


### TERM:

```python
df['term'].value_counts()
```

```python
df['term'] = df['term'].apply(lambda term : int(term[:3]))
```

### GRADE:

**As grade is part of sub_grade, so dropping the grade feature.**

```python
df.drop('grade', axis = 1 , inplace = True)

# or use 
# df = df['grade', axis = 1], if not want to use ---> inplace 
```

**Converting the subgrade into dummy variables. Then concatenate these new columns to the original dataframe, then dropping the original categorical column:**

#### üìå  `drop_first = True`:
1. Using `drop_first=True` in `pd.get_dummies()` to avoid multicollinearity (dummy variable trap) in linear models.
2. This is important for models like **Linear Regression** and **Logistic Regression**.
3. It is **not required** for tree-based models such as **Random Forest** or **XGBoost**.

```python
dummies_subgrade = pd.get_dummies(df['sub_grade'], drop_first = True).astype(int)
df = pd.concat([df , dummies_subgrade], axis = 1)
df.drop('sub_grade', axis = 1 , inplace = True)
```

### VERIFICATION STATUS , APPLICATION TYPE , INITIAL LIST STATUS , PURPOSE:
**Converting these columns: ['verification_status', 'application_type','initial_list_status','purpose'] into dummy variables and concatenate them with the original dataframe, then Dropping original categorical columns:**

```python
# Create dummy variables (one-hot encoded) for 'verification_status', drop first to avoid multicollinearity
dummies_verf_stat = pd.get_dummies(df['verification_status'], drop_first=True).astype(int)

# Create dummy variables for 'application_type', drop first category
dummies_appl_type = pd.get_dummies(df['application_type'], drop_first=True).astype(int)

# Create dummy variables for 'initial_list_status', drop first
dummies_init_lstat = pd.get_dummies(df['initial_list_status'], drop_first=True).astype(int)

# Create dummy variables for 'purpose', drop first to avoid dummy variable trap
dummies_purpose = pd.get_dummies(df['purpose'], drop_first=True).astype(int)

# Concatenate all dummy DataFrames with the main DataFrame along columns (axis=1)
df = pd.concat([df, dummies_verf_stat, dummies_appl_type, dummies_init_lstat, dummies_purpose], axis=1)

# Drop original categorical columns after encoding
df.drop(['verification_status', 'application_type', 'initial_list_status', 'purpose'], axis=1, inplace=True)
```

### HOME OWNERSHIP:
**Reviewing the value_counts for the home_ownership column:**

```python
df['home_ownership'].value_counts()
```

**Converting these to dummy variables. Before which replacing the NONE and ANY with OTHER, though, ending up with just 4 categories, MORTGAGE, RENT, OWN, OTHER. Then concatenating them with the original dataframe, and lastly dropping the original column:**

# üß† Creating the Model:

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()

model.add(Dense(80, activation = 'relu'))
model.add(Dropout(0.4))


model.add(Dense(64 , activation = 'relu'))
model.add(Dropout(0.4))

model.add(Dense(32 , activation = 'relu'))
model.add(Dropout(0.4))

model.add(Dense(16 , activation = 'relu'))
model.add(Dropout(0.4))


model.add(Dense(1 , activation = 'sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
# Declaring EarlyStopping to prevent overfitting
early_stop = EarlyStopping(
    monitor='val_loss',          #  Watch the validation loss during training
    patience=25,                 #  If no improvement for 25 epochs, stop training
    restore_best_weights=True    #  Revert to the model state with the best val_loss
)
```

```python
model.fit(x = x_train ,
          y = y_train ,
          epochs = 400 ,
          batch_size = 64 ,
          validation_data = (x_test, y_test) ,
          callbacks = [early_stop]
          )
```
